```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the Iris Dataset
# scikit-learn provides a built-in function to load the Iris dataset
iris = load_iris()
X = iris.data  # Features (sepal length, sepal width, petal length, petal width)
y = iris.target # Target (species: 0-setosa, 1-versicolor, 2-virginica)

# You can also load it into a Pandas DataFrame for better visualization and understanding
df = pd.DataFrame(data=X, columns=iris.feature_names)
df['species'] = y
df['species_names'] = df['species'].apply(lambda x: iris.target_names[x])

print("--- Iris Dataset Head ---")
print(df.head())
print("\n--- Iris Dataset Description ---")
print(df.describe())
print("\n--- Iris Dataset Species Counts ---")
print(df['species_names'].value_counts())

# 2. Split the Dataset into Training and Testing Sets
# We use 70% of the data for training and 30% for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
# random_state for reproducibility
# stratify=y ensures that the proportion of target classes is the same in both train and test sets

print(f"\nTraining set size: {X_train.shape[0]} samples")
print(f"Testing set size: {X_test.shape[0]} samples")

# 3. Choose and Train a Model (K-Nearest Neighbors Classifier)
# KNN is a simple yet effective algorithm for this dataset
knn = KNeighborsClassifier(n_neighbors=3) # You can experiment with different 'n_neighbors' values
knn.fit(X_train, y_train)

print("\n--- Model Training Complete ---")

# 4. Make Predictions on the Test Set
y_pred = knn.predict(X_test)

# 5. Evaluate the Model
print("\n--- Model Evaluation ---")

# Accuracy Score
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# Classification Report
# Provides precision, recall, f1-score for each class
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))

# Confusion Matrix
# Shows the number of correct and incorrect predictions for each class
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)

# Visualize the Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()

# 6. (Optional) Visualize the Dataset
# Pairplot to see relationships between features and how they separate classes
print("\n--- Visualizing the Dataset (Pairplot) ---")
sns.pairplot(df, hue='species_names', palette='viridis')
plt.suptitle('Pair Plot of Iris Dataset Features by Species', y=1.02)
plt.show()

# 7. (Optional) Make a prediction on new data
print("\n--- Making a Prediction on New Data ---")
# Let's say we have a new iris flower with these measurements:
# sepal length=5.1, sepal width=3.5, petal length=1.4, petal width=0.2 (Looks like a Setosa)
new_data = [[5.1, 3.5, 1.4, 0.2]]
prediction = knn.predict(new_data)
predicted_species = iris.target_names[prediction[0]]
print(f"New data: {new_data}")
print(f"Predicted species: {predicted_species}")

# Another example:
# sepal length=6.0, sepal width=3.0, petal length=4.5, petal width=1.5 (Looks like a Versicolor)
new_data_2 = [[6.0, 3.0, 4.5, 1.5]]
prediction_2 = knn.predict(new_data_2)
predicted_species_2 = iris.target_names[prediction_2[0]]
print(f"New data: {new_data_2}")
print(f"Predicted species: {predicted_species_2}")
```

```python
# 1. Load the Iris Dataset
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
X = iris.data  # Features (sepal length, sepal width, petal length, petal width)
y = iris.target # Target (species: 0-setosa, 1-versicolor, 2-virginica)

df = pd.DataFrame(data=X, columns=iris.feature_names)
df['species'] = y
df['species_names'] = df['species'].apply(lambda x: iris.target_names[x])

print("--- Iris Dataset Head ---")
print(df.head())
print("\n--- Iris Dataset Description ---")
print(df.describe())
print("\n--- Iris Dataset Species Counts ---")
print(df['species_names'].value_counts())
```

```python
# 2. Split the Dataset into Training and Testing Sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(f"\nTraining set size: {X_train.shape[0]} samples")
print(f"Testing set size: {X_test.shape[0]} samples")
```

```python
# 3. Choose and Train a Model (K-Nearest Neighbors Classifier)
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

print("\n--- Model Training Complete ---")
```

```python
# 4. Make Predictions on the Test Set
y_pred = knn.predict(X_test)
```

```python
# 5. Evaluate the Model
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

print("\n--- Model Evaluation ---")

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))

cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
